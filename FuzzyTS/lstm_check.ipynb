```python
from tensorflow.keras import models, layers, preprocessing as kprocessing
import sys, traceback
import numpy as np
import pandas as pd
import os
import itertools
```

```python
df = pd.read_csv("data/deneme.csv",encoding='latin1') 
df = df.drop('data', 1)
df
```

```python

train_size = 5
train_data, test_data = df[0:train_size], df[train_size:]
train_data
```

```python
X_train = train_data.iloc[:, 0:12]
Y_train = train_data.iloc[:, 12:24]
X_test = test_data.iloc[:, 0:12]
Y_test = test_data.iloc[:, 12:24]
X_train.shape
Y_train.shape
```

```python
x_train = X_train.to_numpy()
x_train = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
y_train = Y_train.to_numpy()
y_train = np.reshape(y_train, (y_train.shape[0], 1, y_train.shape[1]))

x_test = X_test.to_numpy()
x_test = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))

y_test = Y_test.to_numpy()
y_test = np.reshape(y_test, (y_test.shape[0], 1, y_test.shape[1]))
print(x_train.shape)
print(y_train.shape)
```

```python

u = 12
lstm_model = models.Sequential()

lstm_model.add(layers.LSTM(units=u, return_sequences = True, activation='tanh', input_shape=(1, u)))
lstm_model.add(layers.Dropout(0.2))
#print("2")
lstm_model.add(layers.LSTM(units=u, return_sequences = True))
lstm_model.add(layers.Dropout(0.2))

lstm_model.add(layers.Dense(u))
lstm_model.compile(optimizer='adam', loss='mse')
#print("6")
lstm_model.summary()
```

```python
lstm_model.fit(x_train, y_train, epochs = 10, batch_size = 32)
```

```python
print("Evaluate on test data")
results = lstm_model.evaluate(x_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```
